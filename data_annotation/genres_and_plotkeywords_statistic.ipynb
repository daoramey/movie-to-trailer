```python
def build_list_token(keywords):
  DI_POS_TYPES = {'NN':'n', 'JJ':'a', 'VB':'v', 'RB':'r'} 
  POS_TYPES = list(DI_POS_TYPES.keys())
  MIN_STR_LEN = 3
  RE_VALID = '[a-zA-Z]'
  stopwords = nltk.corpus.stopwords.words('english')
  stemmer = nltk.stem.PorterStemmer()
  lemmatizer = nltk.stem.WordNetLemmatizer()
  def remove_accents(data):
      return ''.join(x for x in unicodedata.normalize('NFKD', data) if x in string.ascii_letters or x == " ")

  # Process all quotes
  li_tokens = []
  li_token_lists = []
  li_lem_strings = []

  for i,text in enumerate(keywords):
      tokens = [word.lower() for sent in nltk.sent_tokenize(text) for word in nltk.word_tokenize(sent)]
      li_tokens_quote = []
      li_tokens_quote_lem = []
      for token in tokens:
          t = remove_accents(token)
          t = str(t).translate(string.punctuation)
          li_tokens_quote.append(t)
          li_tokens_quote_lem.append("-")
          if t not in stopwords:
              if re.search(RE_VALID, t):
                  if len(t) >= MIN_STR_LEN:
                      pos = nltk.pos_tag([t])[0][1][:2]
                      pos2 = 'n'
                      if pos in DI_POS_TYPES:
                        pos2 = DI_POS_TYPES[pos]
                      
                      stem = stemmer.stem(t)
                      lem = lemmatizer.lemmatize(t, pos=pos2)
                      
                      if pos in POS_TYPES:
                          li_tokens.append((t, stem, lem, pos))
                          li_tokens_quote_lem = li_tokens_quote_lem[:-1] 
                          li_tokens_quote_lem.append(lem)
      li_token_lists.append(li_tokens_quote)
      str_li_tokens_quote_lem = ' '.join(li_tokens_quote_lem)
      li_lem_strings.append(str_li_tokens_quote_lem)
  return li_tokens
```

```python
def count_words(li_tokens):
  df_all_words = pd.DataFrame(li_tokens, columns=['token', 'stem', 'lem', 'pos'])
  df_all_words['counts'] = df_all_words.groupby(['lem'])['lem'].transform('count')
  df_all_words = df_all_words.sort_values(by=['counts', 'lem'], ascending=[False, True]).reset_index()
  df_words = df_all_words.groupby('lem').first().sort_values(by='counts', ascending=False).reset_index()
  return df_words
```

```python
def split_text(text):
  for ch in ["[","]","'"]:
    text = text.replace(ch,"")
  return text.split(",")
```

```python
def text_to_number(texts):
  genres_dict = {
      'Action':0,
      'Adventure':1,
      'Animation':2,
      'Comedy':3,
      'Crime':4,
      'Documentary':5,
      'Drama':6,
      'Fantasy':7,
      'History':8,
      'Horror':9,
      'Western':10,
      'Thriller':11,
      'ScienceFiction':12,
      'Family':13,
      'Music':14,
      'Mystery':15,
      'Romance':16,
      'TVMovie':17,
      'War':18,
      'love':19,
      'murder':20,
      'child':21,
      'school':22,
      'drug':23,
      'violence':24,
      'police':25,
      'lgbt':26,
      'superhero':27,
      'friendship':28,
      'death':29,
      'killer':30,
      'city':31,
      'holiday':32,
      'sport':33,
      'revenge':34,
      'car':35,
      'teenager':36,
      'alien':37,
      'prison':38,
      'dog':39,
      'kidnapping':40,
      'space':41,
      'wedding':42,
      'ghost':43,
      'monster':44,
      'zombie':45,
      'gang':46,
      'blood':47,
      'supernatural':48,
      'dystopia':49,
      'detective':50,
      'vampire':51,
      '':52
  }
  nums = []
  for text in texts:
    num = genres_dict[text]
    nums.append(num)
  return nums

def label(num):
  g_type = 52
  g_label = np.zeros(g_type+1)
  for n in num:
    g_label[n] = 1
  return g_label[:g_type]

def rp(text):
 for ch in ["[","]","'"," ","''"]:
   text = text.replace(ch,"")
 return text.split(",")
```

```python
def text_to_number(texts):
  genres_dict = {
      'Action':0,
      'Adventure':1,
      'Animation':2,
      'Comedy':3,
      'Crime':4,
      'Documentary':5,
      'Drama':6,
      'Fantasy':7,
      'History':8,
      'Horror':9,
      'Western':10,
      'Thriller':11,
      'ScienceFiction':12,
      'Family':13,
      'Music':14,
      'Mystery':15,
      'Romance':16,
      'TVMovie':17,
      'War':18,
      '':19
  }
  nums = []
  for text in texts:
    num = genres_dict[text]
    nums.append(num)
  return nums

def rp(text):
 for ch in ["[","]","'"," "]:
   text = text.replace(ch,"")
 return text.split(",")

def label(g):
  g_type = 19
  t_num = 21000
  g_label = np.zeros((t_num,g_type+1))
  for i in range(t_num):
    for j in g[i]:
      g_label[i][j] = 1
  return g_label

def convert_genres_to_label(genres):
  g = []
  for i in range(len(genres)):
    genre = rp(genres[i])
    genre = text_to_number(genre)
    g.append(genre)
  g_label = label(g)
  return g_label

def main(metadata):
  genres = metadata['genres']
  id = metadata['id']
  g_label = convert_genres_to_label(genres)
  df = pd.DataFrame(data = g_label[:,:19], index = id, columns = ['Action','Adventure','Animation','Comedy','Crime','Documentary','Drama','Fantasy','History','Horror','Western','Thriller','ScienceFiction','Family','Music','Mystery','Romance','TVMovie','War'])
  return df
```

```python
import os
import numpy as np
import tensorflow as tf
import h5py

def load_data_from_h5py(data_path):
  f1 = h5py.File(data_path,'r')
  y = np.array(f1["y"][:])
  return y

data_path = '../model/CNN-based_Visual_Model/Training_Dataset/13K_training_dataset(keyword)/trainingdata_y.hdf5'
y = load_data_from_h5py(data_path)
print(y.shape)
```
